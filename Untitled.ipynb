```python
df.columns
```

### GENDER

```python
ax = sns.countplot(data = df, x = 'gender' )

for bars in ax.containers:
    ax.bar_label(bars)
```

### HOW MANY STUDENTS TAKES PREPARATORY CLASSES

```python
count_test = df['test preparation course'].value_counts()
labels = df['test preparation course'].value_counts().index
plt.figure(figsize= (6,6))
plt.pie(count_test,labels=labels,autopct='%1.1f%%')
plt.legend(labels)
plt.show()
```

### SECTIONS(GROUPS)

```python
ax = sns.countplot(data = df, x = 'race/ethnicity' , hue = 'gender')

for bars in ax.containers:
    ax.bar_label(bars)
```
